# Finance Data Project

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
from pandas_datareader import data, wb
import datetime
import seaborn as sns
```

# Data

```python
start=datetime.datetime(2006,1,1)
end=datetime.datetime(2020,6,30)
source="yahoo"
```

```python
# Bank of America
BAC=data.DataReader("BAC",source,start,end)
# Citigroup
C=data.DataReader("C",source,start,end)
# Goldman Sachs
GS=data.DataReader("GS",source,start,end)
# JP Morgan
JPM=data.DataReader("JPM",source,start,end)
# Morgan Stanley
MS=data.DataReader("MS",source,start,end)
# Wells Fargo
WFC=data.DataReader("WFC",source,start,end)

```

```python
BAC
```

```python
tickers=["BAC","C","GS","JPM","MS","WFC"]
```

```python
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
bank_stocks
```

```python
bank_stocks.columns.names=["Bank Ticker","Stock Info"]
```

```python
bank_stocks.head()
```

# Exploratory Data Analysis

```python
# bank_stocks["BAC"]["Close"].max()
for tick in tickers:
    print(bank_stocks[tick]["Close"].max())
```

```python
bank_stocks.xs(key="Close",axis=1,level="Stock Info").max()
```

```python
returns=pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+" Return"]=bank_stocks[tick]["Close"].pct_change(1)
```

```python
returns
```

```python
sns.pairplot(returns)
```

```python
returns.idxmin()
```

```python
returns.idxmax()
```

```python
returns.std()
```

```python
returns.loc["2015-01-01":"2015-12-31"].std()
```

```python
sns.distplot(returns.loc["2015-01-01":"2015-12-31"]["MS Return"],
             color="green",bins=40)
```

```python
sns.distplot(returns.loc["2008-01-01":"2008-12-31"]["C Return"],
             color="red",bins=40)
sns.distplot(returns.loc["2019-01-01":"2019-12-31"]["C Return"],
             color="blue",bins=40)
```
