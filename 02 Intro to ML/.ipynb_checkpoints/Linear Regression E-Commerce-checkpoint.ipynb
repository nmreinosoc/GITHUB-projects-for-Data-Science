# Training and Testing Data

```python
customers.columns
```

```python
y=customers["Yearly Amount Spent"]
y
```

```python
X=customers[['Avg. Session Length', 'Time on App',
             'Time on Website', 'Length of Membership']]
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
X_train, x_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
X_train.describe()
```

```python
x_test.describe()
```

```python
y_train.describe()
```

```python
y_test.describe()
```

```python
# Import LinearRegression from sklearn.linear_model 
from sklearn.linear_model import LinearRegression
```

```python
# Create an instance of a LinearRegression() model to lm
lm = LinearRegression()
```

```python
# Train/fit lm on the training data
lm.fit(X_train,y_train)
```

```python
# Print out the coefficients of the model
coefficient= pd.DataFrame(lm.coef_,index=X_train.columns,columns=["Coeff"])
coefficient
```

```python
# Use lm.predict() to predict off the x_test set of the data
predictions=lm.predict(x_test)
predictions
```

```python
# Create a scatterplot of the real test values versus the predicted values
plt.scatter(x=predictions,y=y_test)
plt.xlabel("Y Test (True Values)")
plt.ylabel("Predicted Values")
plt.title("Predictes vs. True Values")
```

```python
pred_ytest=pd.DataFrame([predictions,y_test],index=["Predictions","y_Test"]).T
pred_ytest
```

```python
sns.lmplot(x="Predictions",y="y_Test",data=pred_ytest)
```

# Evaluating the Model

```python
# Calculate the mean absolute square, mean squared error, and root mean squared error
import sklearn.metrics
```

```python
print("MAE: ",sklearn.metrics.mean_absolute_error(y_test,predictions))
```

```python
print("MSE: ",sklearn.metrics.mean_squared_error(y_test,predictions))
```

```python
print("RMSE: ",np.sqrt(sklearn.metrics.mean_squared_error(y_test,predictions)))
```

```python
print("R^2: ",sklearn.metrics.explained_variance_score(y_test,predictions))
```
