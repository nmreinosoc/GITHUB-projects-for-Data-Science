```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn")
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer.DESCR)
```

```python
df = pd.DataFrame(cancer["data"],columns=cancer["feature_names"])
df
```

```python
df.corr()
```

```python
print(cancer["target"])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
```

```python
scaler_stan = StandardScaler()
normalize = normalize(df)
```

```python
scaler_stan.fit(df)
```

```python
scaled_data = scaler_stan.transform(df)
```

```python
# PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca_2 = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
pca_2.fit(normalize)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
x_pca_2=pca_2.transform(normalize)
```

```python
x_pca_2.shape
```

```python
scaled_data.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer["target"],cmap="rainbow")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca_2[:,0],x_pca_2[:,1],c=cancer["target"],cmap="rainbow")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer["feature_names"])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap="plasma")
```
